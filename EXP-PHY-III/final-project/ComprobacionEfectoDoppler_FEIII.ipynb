```python
#radio (R) , vangular (w)  , distancia cntro-objeto (D) , velocidad sonido (v)
#frecuancia emita (f) , frecuencia percibida(fp) , tiempo (t)
def efecto_doppler(R,D,w,v,f):
  epsilon = R/D
  t = np.linspace(0,4*np.pi/w,1000)
  fp = f* ( 1- ( (R*w/v) * np.sin(w*t) * (1/( 1+ epsilon**2 - 2*epsilon*np.cos(w*t) )**1/2) )  )

  fig , ax = plt.subplots(figsize=(15,5))
  ax.plot(t,fp , color = "r")
  ax.set_title("Efecto Doppler en un Mvto Circular Uniforme ", fontsize=14, fontweight="bold")
  ax.set_xlabel("tiempo (s)", fontsize=12)
  ax.set_ylabel("Frecuencia percibida (Hz)", fontsize=12)
  ax.legend(["Frecuencia Doppler"])
  ax.grid(True, linestyle="--", alpha=0.6)
  plt.show()
  return
```

```python
#radio (R) , vangular (w)  , distancia cntro-objeto (D) , velocidad sonido (v)
#frecuancia emita (f) , frecuencia percibida(fp) , tiempo (t)
def efecto_doppler(w,f,t):
  R = 0.14606
  D = R + 0.00413
  v = 343
  epsilon = R/D
  fp = f* ( 1- ( (R*w/v) * np.sin(w*t) * (1/( 1+ epsilon**2 - 2*epsilon*np.cos(w*t) )**1/2) )  )
  return fp
```

```python
import numpy as np
import matplotlib.pyplot as plt

def efecto_doppler_receptor_mcu(R, D, w, v, f):
    epsilon = R / D  # Relación entre radio de giro y distancia al emisor
    t = np.linspace(0, 4*np.pi/w, 1000)  # Dos ciclos completos de la trayectoria

    # Cálculo de la frecuencia percibida
    fp = f * (1 + ((R*w/v) * np.sin(w*t) / np.sqrt(1 + epsilon**2 - 2*epsilon*np.cos(w*t))))

    # Graficar
    fig, ax = plt.subplots(figsize=(15,5))
    ax.plot(t, fp, color="b", label="Frecuencia Percibida")

    # Estilización de la gráfica
    ax.set_title("Efecto Doppler con Receptor en MCU", fontsize=14, fontweight="bold")
    ax.set_xlabel("Tiempo (s)", fontsize=12)
    ax.set_ylabel("Frecuencia percibida (Hz)", fontsize=12)
    ax.legend()
    ax.grid(True, linestyle="--", alpha=0.6)
    plt.show()

# Ejemplo de uso
efecto_doppler_receptor_mcu(R=10, D=20, w=2*np.pi, v=343, f=900)
```
