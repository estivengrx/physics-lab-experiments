```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from math import factorial
```

```python
datos_geiger = pd.read_csv('/content/drive/MyDrive/Experimental III/Distribuciones/datos-geiger-19-02-2025 - Datos extraidos.csv')
datos_geiger
```

```python
# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Histograma con etiquetas y mejoras
ax.hist(datos_geiger, bins=144, color='purple', edgecolor='black', alpha=0.75)
ax.set_title("Distribución de detecciones del Geiger", fontsize=14, fontweight="bold")
ax.set_xlabel("Número de eventos detectados", fontsize=12)
ax.set_ylabel("Frecuencia", fontsize=12)
ax.grid(True, linestyle="--", alpha=0.6)

# Mostrar la gráfica
plt.show()
```

```python
N = 144
p = np.mean(datos_geiger)

mu = N * p

for i in range(0, 9):
  f_poisson = ((mu ** i ) * np.exp(-mu)) / factorial(i)
  print(f'La probabilidad de que se detecten {i} partículas en un intervalor de 10s es de {f_poisson}')
```
