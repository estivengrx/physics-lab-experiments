# Punto 3

3. Utilice la ecuación del punto 1 para determinar
la aceleración de la gravedad para cada periodo. Agrupe las aceleraciones en
grupos de 5, 15 y 30 datos y realice los siguientes pasos con cada grupo: (a)
determine la aceleración promedio y el error estándar, (b) En una misma figura
grafique el valor promedio con barras de error de un error estándar, 3 errores
estándar, y con la corrección de t-student para una incerteza del 0.27%.
Incluya en la gráfica como valor aceptado para la aceleración g=977 cm/s^2.
Analice y concluya sobre el resultado para cada grupo y cada error reportado.

```python
from scipy.stats import t

longitud_constante = 0.668 # Longitud constante del péndulo en metros

def calcular_gravedad(periodo, longitud):
    # Calcular la gravedad
    gravedad = (4 * np.pi**2 * longitud) / periodo**2
    return gravedad

def tests_gravedad(datos, grupos_de_datos=[5, 15, 30]):
    g_aceptado = 977  # cm/s^2
    g_aceptado_m_s2 = g_aceptado / 100  # Convertir a m/s^2

    fig, axs = plt.subplots(len(grupos_de_datos), 1, figsize=(10, 15))
    fig.suptitle('Gravedad calculada con diferentes tamaños de muestra y barras de error')

    for i, grupo in enumerate(grupos_de_datos):
        # Seleccionando de forma al azar el numero de datos
        datos_grupo = datos.sample(n=grupo)
        # print(datos_grupo)
        # Calculando la gravedad
        gravedad = calcular_gravedad(datos_grupo['periodo(segundos)'], longitud_constante)
        aceleracion_promedio = gravedad.mean()
        error_estandar = gravedad.std() / np.sqrt(grupo)

        # Calcular la corrección de t-student para una incerteza del 0.27%
        t_value = t.ppf(1 - 0.0027 / 2, df=grupo - 1)
        error_t_student = t_value * error_estandar

        # Graficar
        axs[i].errorbar(1, aceleracion_promedio, yerr=error_estandar, fmt='o', label='1 Error estándar')
        axs[i].errorbar(2, aceleracion_promedio, yerr=3 * error_estandar, fmt='o', label='3 Errores estándar')
        axs[i].errorbar(3, aceleracion_promedio, yerr=error_t_student, fmt='o', label='Corrección t-student (0.27%)')
        axs[i].axhline(y=g_aceptado_m_s2, color='r', linestyle='--', label='g aceptado (9.77 m/s^2)')
        axs[i].set_title(f'Grupo de {grupo} datos')
        axs[i].set_xticks([1, 2, 3])
        axs[i].set_xticklabels(['1 Error estándar', '3 Errores estándar', 'Corrección t-student'])
        axs[i].set_ylabel('Gravedad (m/s^2)')
        axs[i].legend()
    fig.text(0.5, 0.04, 'Nota: Las gráficas no empiezan en 0 para mostrar mejor los cambios de las barras de error', ha='center')
    plt.xlabel('Tipo de barra de error')
    plt.tight_layout(rect=[0, 0.08, 1.5, 0.98])
    plt.show()
```

```python
print("Análisis de gravedad para datos del cronómetro")
tests_gravedad(datos_cronometro_filtrados)
```

```python
print("Análisis de gravedad para datos de la fotocompuerta")
tests_gravedad(datos_fotocompuerta)
```

#Punto 4
4. Determine el porcentaje de error del valor
 central para cada grupo con respecto al valor esperado para Medellín. Analice y
 concluya.

```python
def porcentaje_error(valor_medido, valor_teorico):
    return abs((valor_medido - valor_teorico) / valor_teorico) * 100

def analizar_grupos(datos, grupos_de_datos=[5, 10, 15]):
    g_aceptado = 977  # cm/s^2
    g_aceptado_m_s2 = g_aceptado / 100  # Convertir a m/s^2

    resultados = []

    for grupo in grupos_de_datos:
        datos_grupo = datos.sample(n=grupo)  # Seleccionar datos al azar
        gravedad = calcular_gravedad(datos_grupo['periodo(segundos)'], longitud_constante)

        # Calcular la media y el porcentaje de error con 2 cifras significativas
        media = "{:.3g}".format(gravedad.mean())
        error_media = "{:.3g}".format(porcentaje_error(float(media), g_aceptado_m_s2))


        resultados.append({
            "Tamaño del grupo": grupo,
            "Media (m/s²)": media,
            "Error Media (%)": error_media
        })

    df_resultados = pd.DataFrame(resultados)
    print(df_resultados.to_string(index=False))
    return df_resultados



print("Resultados con datos del cronómetro:")
df_cronometro = analizar_grupos(datos_cronometro)

print("\nResultados con datos de la fotocompuerta:")
df_fotocompuerta = analizar_grupos(datos_fotocompuerta)
```

Analizando los datos obtenidos, se evidencia una clara mejora en el porcentaje de error del valor central registrado con la fotocompuerta en comparación con el obtenido mediante el cronómetro.

```python
def mejor_resultado(df_resultados):
    # Convertir los valores de la columna "Error Media (%)" a float para hacer comparaciones
    df_resultados["Error Media (%)"] = df_resultados["Error Media (%)"].astype(float)

    # Encontrar la fila con el menor error porcentual
    mejor_fila = df_resultados.loc[df_resultados["Error Media (%)"].idxmin()]


    print("\n Mejor resultado obtenido:")
    print(f"Tamaño del grupo: {int(mejor_fila['Tamaño del grupo'])}")
    print(f"Aceleración Media (m/s²): {mejor_fila['Media (m/s²)']}")
    print(f"Error porcentual: {mejor_fila['Error Media (%)']}%")
```
